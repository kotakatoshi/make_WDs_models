```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# ファイルパスを指定
# ファイルパスを指定

file_path = "LOGS/history.data"

# ファイルを読み込み、5行目をヘッダーとして指定
df = pd.read_csv(file_path, delim_whitespace=True, skiprows=5)


# 時間の計算
df["time"] = df["star_age"] - df["star_age"].min()
time = df["time"].to_numpy()
# 各物理量を取得
R= 10 ** df["log_R"].to_numpy() 
Rho_c = 10 ** df["log_center_Rho"].to_numpy() 
cntr_temp = 10 ** df["log_center_T"].to_numpy() 
one_core_radius = df["one_core_radius"].to_numpy()
L_nuc= 10 ** df["log_Lnuc"].to_numpy() 
L = 10 ** df["log_L"].to_numpy()  * 3.8e33  # 太陽光度でスケーリング
model = df["model_number"].to_numpy()
center_entropy = df["center_entropy"].to_numpy()
```
